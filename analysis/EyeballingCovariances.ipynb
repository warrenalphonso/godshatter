```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(cov, mean=[0, 0], n=1000):
    # Generate the random 2D points
    x, y = np.random.multivariate_normal(mean, cov, n).T

    # Plot the points
    plt.figure(figsize=(6, 6))
    plt.scatter(x, y, s=10)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.axis("equal")  # Equal aspect ratio ensures that the x and y scales are identical
    plt.show()
```

```python
cov = [[16, 0], [0, 25]]
plot(cov)
```

```python
cov = [[36, 16], [16, 64]]
plot(cov)
```

```python
cov = [[64, -40], [-40, 36]]
plot(cov)
```

```python
cov = [[100, 0], [0, 1]]
plot(cov)
```

```python
cov = [[1, 0], [0, 5]]
plot(cov)
```
